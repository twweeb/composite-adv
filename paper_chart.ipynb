```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.font_manager as fm
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'cm'
myfont = fm.FontProperties(fname=r'/home/hsiung/labs/NimbusRomNo9L-Reg.otf', size=16) # Font Setting
```

```python
labels = ['Clean (accuracy)', 'Hue', 'Hue + Sat.',
          'Hue + Sat. + Rot.', 'Hue + Sat. + Rot.\n+ Bri. + Con.',
          'Hue + Sat. + $\ell_\infty$',
          'Hue + Sat. + Rot.\n+ Bri. + Con. + $\ell_\infty$']
# font = {'family': 'Linux Libertine',}
data = [[0.761299968, 0.340470254,0.3979771137,0.5835544467,0.748771846,0.999842375,0.999973729],
[0.624239981,0.391996682,0.448737681,0.653722942,0.8566256613,0.7678457052,0.9498269893],
[0.5996,0.2132230401,0.2409099936,0.3108612895,0.3961238265,0.7050170004,0.8056908399]]

data = [[round(prec*100,2) for prec in item] for item in data]
X = np.arange(len(labels))
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(15,5))
stand_train = ax.bar(X - 0.25, data[0], color='gray', width=width, label='Standard')
lind_rob    = ax.bar(X, data[1], color='orange', width=width, label='$\ell_\infty$ Robust')
gat         = ax.bar(X + 0.25, data[2], color='#17c0bd', width=width, label='GAT (Proposed)')

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel('Attack Success Rate', fontsize=14)
# ax.set_title('Scores by group and gender')
ax.set_xticks(X)
ax.set_xticklabels(labels)
plt.xticks(fontsize=14, fontproperties=myfont)
plt.ticklabel_format(axis='y', useMathText=True)

ax.legend(loc='best',bbox_to_anchor=(0.205,1.1), prop=myfont)

ax.get_yaxis().set_visible(False)
# ax.get_xaxis().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.grid(False)

ax.bar_label(stand_train, fmt='%g', padding=3)
ax.bar_label(lind_rob, padding=3)
ax.bar_label(gat, padding=3)

fig.tight_layout()

plt.savefig("images/header-chart.pdf")
plt.show()
```
